```python
# rename columns

new_column_names = ['stop_' + str(i) for i in range(41)]
```

```python
raw.columns = new_column_names
```

```python
# rename first column

raw.rename(columns={'stop_0': 'dep_stn'}, inplace=True)
```

```python
raw.head()
```

```python
# check if there are duplicates by concatenate all the stations of a row and then check for duplicates. each raw indicated the starts and end of a train route

def concatenate_non_nan(row):
    return ', '.join([str(value) for value in row if not pd.isna(value)])
```

```python
raw['train_route'] = raw.apply(concatenate_non_nan, axis=1)
```

```python
raw.head()
```

```python
#remove duplicates values
raw = raw.drop_duplicates(subset='train_route')
```

```python
raw.info()
```

```python
# adding a column that shows the number of stations conneted by each route

def count_non_null(row):
    return row.iloc[:41].count()
```

```python
raw['stn_connected'] = raw.apply(count_non_null, axis=1)
```

```python
raw.head()
```

```python
raw.info()
```

```python
raw[raw['stn_connected'] == 1]
```

```python
# remove rows where only one station is displayed

raw = raw[raw['stn_connected'] != 1]
```

```python
raw.info()
```

```python
raw['dep_stn'].isna().sum()
```

```python
# remove the null value in dep_stn column

raw = raw.dropna(subset=['dep_stn'])
```
